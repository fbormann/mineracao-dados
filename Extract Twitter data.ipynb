```python
import twitter
import os
import pandas
```

If you want to learn how to get the credentials for twitter, please refer to this docs from the python-twitter package.
https://python-twitter.readthedocs.io/en/latest/getting_started.html

```python
secrets_path = "secrets/secrets.txt"

count_line = 1
secrets_data = {}
with open(secrets_path, 'r') as fp:
    for line in fp:
        key, value = line.split("=")
        secrets_data[key] = value.rstrip()
        count_line +=1
```

```python
secrets_path
```

```python
api_crawler = twitter.Api(consumer_key=secrets_data["API_KEY"], 
                          consumer_secret=secrets_data["API_SECRET"],
                          access_token_key=secrets_data["ACCESS_TOKEN"],
                          access_token_secret=secrets_data["TOKEN_SECRET"]
                         )
```

```python
source_data = "twitter_data/tweets_labeled.txt"
df = pandas.read_csv(source_data, names=["tweet_id", "label"])
```

```python
df.head(10)
```

```python
df.iloc[200:100]
```

```python
bulk_size = 100
pos = 0
data = []
while pos < df.shape[0]:
    bulk_ids = df.iloc[pos:(pos+bulk_size)]["tweet_id"].tolist()
    pos += bulk_size
    data.append(api_crawler.GetStatuses(bulk_ids))
```

```python
len(data)
```

```python
data[0][0].__dict__
```

```python
dataset = []
for bulk_list in data:
    for datum in bulk_list:
        dataset.append(datum.__dict__)
```

```python
df_tweet_info = pandas.DataFrame(dataset)
```

```python
df_tweet_info.head(10)
```

```python
df_tweet_info.to_csv("twitter_data/labeled_tweets_info.csv")
```
