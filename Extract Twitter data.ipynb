```python
secrets_path = "secrets/secrets.txt"

count_line = 1
secrets_data = {}
with open(secrets_path, 'r') as fp:
    for line in fp:
        key, value = line.split("=")
        secrets_data[key] = value.rstrip()
        count_line +=1
```

```python
print(secrets_data)
```

```python
api_crawler = twitter.Api(consumer_key=secrets_data["API_KEY"], 
                          consumer_secret=secrets_data["API_SECRET"],
                          access_token_key=secrets_data["ACCESS_TOKEN"],
                          access_token_secret=secrets_data["TOKEN_SECRET"]
                         )
```
